```python
import pandas as pd
import numpy as np
df_logret = pd.read_csv('CryptoLogReturnsData.csv')
df_logret = df_logret.drop(['trade_date'], axis=1)
df_logret = df_logret.dropna()
df_logret_abs = df_logret.abs()
df_logret.head()
```

```python
df_logret.plot(figsize=(15, 10)).axhline(color='black', linewidth=2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Compute the correlation matrix
corr = df_logret.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(30, 30))
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, fmt = '.4f', mask=mask, center=0, square=True, linewidths=.5)
```
